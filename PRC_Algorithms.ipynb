# Adding Unoccupied Hours as Zeros

During unoccupied hours, there is no occupant in the space, thus there is no user feedback due to visual discomfort. 
So, we assume ZERO values for unoccupied hours that need to be repeated before/after working hours (e.g.8am-5pm) throughout all available columns in the data frame (2304 columns in this case)

```python
cols = list(alldata.columns)
new_chunks = np.split(alldata, 365) #dividing dataframe into days
new_df = pd.DataFrame()
for i in range(len(new_chunks)): #Adding additional values during unoccupied hours
    part1 = np.zeros((24, 2304))
    part2 = new_chunks[i].values
    part3 = np.zeros((20, 2304))
    one_day_data = np.vstack((part1, part2, part3))
    one_day_data = pd.DataFrame(one_day_data)
    new_df = pd.concat([new_df, one_day_data])
```

# Integrating PnCn Loops into a Global Loop

This is the CORE of the PRC algorithm. This algorithm is compatible to different adjustments that in my research, I altered the algorithm and studied the results based on three extra scenarios:
+Limiting user preferences extremes (maximum and minimum values among users) by 30% 
+Equalizing the user preferences (assuming all users have similar visual comfort preferences)
+Re-ordering the prioritization list (originally Ev is prior to Et and VR, but in this scenario, Et is prior to VR and Ev)

Final control decisions are based on several drivers and their respective definitions as a proof of concept:
•	P1C1 – adjust the shading position in response to acceptable EV thresholds for corresponding user with least impact on neighbors’ view ratio,
•	P1C2 – Keep the shades in current position,
•	P1C3 – Changing to fully-closed shading position for all columns,
•	P2C1 – Adjust the shading position in response to view ratio where EV is in comfortable range for all other users,
•	P2C2 – Keep the shades in current position,
•	P2C3 - Adjust the shading position in response to view ratio where EV is in comfortable range for all other users if possible, (there might be times during a day that providing sufficient view can increase risk of discomfort glare)
•	P3C1 - Adjust the shading position in response to task illuminance where EV is in comfortable range,
•	P3C2 – Keep the shades in current position,
•	P3C3 – Changing shading position corresponding to the same column to fully-open position.
•	‘NoAction’ – When both users of the same column are absent, position the shades according to the required indoor illuminance for each column (Stage B.2), only if there is no other feedback from other users.

NOTE: The algorithm can be optimized in future investigations or can host AI to enhance its performance.
